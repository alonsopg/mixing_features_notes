# Mixing Feats notes

Quick notes of how to mix feature extractors with scikit learn. Let's read the data:

```python
#Read the .csv file
import pandas as pd
df = pd.read_csv('/Users/user/Tesis/escritorio_anterior/MICAI/todos.csv',
                     header=0, sep=',', names=['id', 'content', 'label'])
df
```

## Import scikit-learn vectorizers:

```python
#vectorizer 1
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vect = TfidfVectorizer(use_idf=True, smooth_idf=True,
                             sublinear_tf=False, ngram_range=(2,2))
#vectorizer 2
from sklearn.feature_extraction.text import CountVectorizer
bow = CountVectorizer(ngram_range=(2,2))

#vectorizer 3
from sklearn.feature_extraction.text import HashingVectorizer
hash_vect = HashingVectorizer(ngram_range=(2,2))
```

Combine the above vectorizers in a single feature matrix, using FeatureUnion interface:

```python
from sklearn.pipeline import FeatureUnion
combined_features = FeatureUnion([
                                  ("tfidf_vect", tfidf_vect),
                                  ("bow", bow),
                                  ("hash",hash_vect)
                                ])
```

## Vectorize the data and the labels:

```python
X_combined_features = combined_features.fit_transform(df['content'].values)
y = df['label'].values
```

Let's check the dimension of the document-term matrix and print the sparse matrix:

```python
X_combined_features.shape
```

```python
print '\n',X_combined_features
```

## Let's cross validate, with kfold cv:

```python
from sklearn import cross_validation
kf = cross_validation.KFold(n=len(y),n_folds=10)

for train_index, test_index in kf:
    print "\nTRAIN:\n", train_index, "\n TEST:\n", test_index
    X_train, X_test = X_combined_features[train_index], X_combined_features[test_index]
    y_train, y_test = y[train_index], y[test_index]
```

## Let's classify with an SVM

```python
from sklearn.svm import SVC
svm = SVC(kernel='linear', C = 1)
svm.fit(X_combined_features, y)
estimate = svm.predict(X_test)
print estimate

print '\n quick estimation:\n', svm.fit(X_combined_features[train_index], y[train_index]).score(X_combined_features[test_index], y[test_index])
```

## We can also perform a feature selection from the above matrix:

```python
from sklearn.feature_selection import VarianceThreshold
feature_selection = VarianceThreshold(threshold = (.8 * (1 - .8)))

reduced_training_matrix = feature_selection.fit_transform(X_combined_features)
reduced_training_matrix.toarray()
```

Analogosly with the testing matrix:

```python
reduced_testing_matrix = feature_selection.fit_transform(X_test)
reduced_testing_matrix.toarray()
```
